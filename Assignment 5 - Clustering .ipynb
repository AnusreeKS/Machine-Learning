```python
# Import necessary libraries
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage
```

<span style="color:green; font-weight:bold; font-size:24px">Loading and Preprocessing</span>

```python

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
# Drop the species column (i.e., the labels)
# In the sklearn iris dataset, target is separate, so we simply ignore it
print("Iris dataset loaded and species column excluded.")
```

```python
df
```

<span style="color:green; font-weight:bold; font-size:24px">KMeans Clustering</span>


 <span style="color:blue; font-weight:bold; font-size:16px">KMeans clustering partitions data into k distinct, non-overlapping clusters based on feature similarity. It initializes 'k' centroids, assigns each point to the nearest centroid, and updates centroids until convergence.</span>
 

  <span style="color:blue; font-weight:bold; font-size:16px">KMeans works well on structured datasets like Iris where natural groupings (e.g., species) are expected.</span>

  <span style="color:orange; font-weight:bold; font-size:16px">Applying the ELBOW Method to find the optimum number of clusters(k).</span>

```python
k_range= range(1,10)
sse=[]
for k in k_range :
    km=KMeans(n_clusters=k)
    km.fit(df[iris.feature_names])
    sse.append(km.inertia_)
sse   #SSE Values to diffrent k
```

```python
plt.xlabel("k")
plt.ylabel("Sum of Squared Errors(SSE)")
plt.plot(k_range,sse)
plt.show()
```

  <span style="color:orange; font-weight:bold; font-size:16px">Optimum no. of clusters (k) =3 from the elbow method.</span>

  <span style="color:blue; font-weight:bold; font-size:20px">Apply KMeans</span>

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['kmeans_cluster'] = kmeans.fit_predict(df[iris.feature_names])
```

```python
df['kmeans_cluster']
```

```python
# Visualize with PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df.drop('kmeans_cluster', axis=1))
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_pca[:, 0], y=df_pca[:, 1], hue=df['kmeans_cluster'], palette='Set1')
plt.title("KMeans Clustering (PCA-reduced)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```

<span style="color:green; font-weight:bold; font-size:24px">Hierarchial Clustering</span>

<span style="color:blue; font-weight:bold; font-size:16px">Hierarchical clustering builds a tree of clusters (dendrogram) either by agglomerating points bottom-up
or dividing them top-down, based on distance metrics.</span>

 <span style="color:blue; font-weight:bold; font-size:16px"> It's useful when the number of clusters is not known in advance, and for understanding hierarchical relationships.</span>

```python
# Apply Agglomerative Clustering
hierarchical = AgglomerativeClustering(n_clusters=3)
df['hierarchical_cluster'] = hierarchical.fit_predict(df.drop('kmeans_cluster', axis=1))
```

```python
# Visualize clusters (PCA)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_pca[:, 0], y=df_pca[:, 1], hue=df['hierarchical_cluster'], palette='Set2')
plt.title("Hierarchical Clustering (PCA-reduced)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```

  <span style="color:blue; font-weight:bold; font-size:20px">Dendrogram</span>

```python

linked = linkage(df.drop(['kmeans_cluster', 'hierarchical_cluster'], axis=1), method='ward')
plt.figure(figsize=(10, 5))
dendrogram(linked, truncate_mode='level', p=3)
plt.title("Dendrogram - Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```
